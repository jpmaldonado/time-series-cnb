```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
flats = pd.read_csv("./data/apartments.csv")
```

```python
flats.head()
```

```python
X=flats.iloc[:,0:5]
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler

X_norm = StandardScaler().fit_transform(X)
```

```python
km = KMeans(n_clusters=5)
```

```python
labels = km.fit_predict(X_norm)
```

```python
flats['cluster'] = labels
```

```python
flats.groupby('cluster')['m2.price'].mean()
```

```python
flats.boxplot('m2.price', by='cluster')
```

```python
plt.scatter(flats['floor'],flats['m2.price'], c=flats['cluster'])
plt.xlabel('Surface area')
plt.ylabel('Price per m2')
plt.show()
```
